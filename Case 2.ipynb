#### **Analisis tentang variabel `GarageArea`**

Analisis Variabel GarageArea

Ukuran Pusat:

* Median (480.0): Setengah dari rumah dalam dataset memiliki luas garasi 480 kaki persegi atau kurang. Ini menunjukkan ukuran garasi yang umum dalam dataset.
* Rata-rata (472.98): Rata-rata luas garasi sedikit lebih rendah dari median, sekitar 472.98 kaki persegi. Ini menunjukkan kemungkinan adanya beberapa rumah dengan garasi yang sangat kecil atau tidak ada garasi sama sekali, yang menarik rata-rata ke bawah.

Variabilitas:

* Standar Deviasi (213.80): Nilai ini menunjukkan rata-rata perbedaan antara luas garasi individual dan rata-rata luas garasi. Standar deviasi sebesar 213.80 menunjukkan tingkat variabilitas yang sedang dalam ukuran garasi. Artinya, terdapat sebaran yang cukup terlihat dalam luas garasi di dalam dataset.


Kemungkinan Interpretasi dan Pertimbangan Lebih Lanjut:

* Distribusi: Fakta bahwa rata-rata sedikit lebih rendah dari median mengisyaratkan kemungkinan adanya sedikit kemiringan negatif (left skew) dalam distribusi luas garasi. Ini berarti bahwa mungkin ada lebih banyak rumah dengan garasi yang lebih kecil.

* Pencilan (Outlier): Standar deviasi menunjukkan adanya beberapa rumah dengan luas garasi yang luar biasa besar atau kecil dibandingkan dengan rata-rata. Perlu dilakukan penyelidikan lebih lanjut terhadap pencilan ini untuk melihat apakah mereka mewakili karakteristik rumah tertentu atau anomali data.

* Hubungan dengan Variabel Lain: Akan menarik untuk mengeksplorasi bagaimana GarageArea berhubungan dengan variabel lain seperti SalePrice atau tipe rumah. Misalnya, apakah rumah dengan garasi yang lebih besar cenderung memiliki harga jual yang lebih tinggi?

* Nilai yang Hilang: Meskipun kode yang diberikan tidak secara khusus mengatasinya, perlu diperhatikan potensi nilai yang hilang dalam kolom GarageArea. Jika ada, mereka harus ditangani dengan tepat sebelum menarik kesimpulan.

Kesimpulan:

Analisis ini menunjukkan bahwa ukuran garasi yang umum dalam dataset adalah sekitar 480 kaki persegi. Namun, terdapat variasi yang cukup dalam ukuran garasi, dengan beberapa rumah memiliki garasi yang jauh lebih besar atau lebih kecil. Eksplorasi lebih lanjut diperlukan untuk memahami distribusi secara lebih detail, mengidentifikasi pencilan potensial, dan menyelidiki hubungan dengan variabel relevan lainnya.

## Pemrosesan Data Variabel Numerikal

```python
# Transformasi log-10 SalePrice
houseLogPrice = np.log10(house['SalePrice'])

### Tulis kode Anda di sini
# Median
medianLogP=houseLogPrice.median()
# Mean
meanLogP=houseLogPrice.mean()
# Std Dev
stdLogP=houseLogPrice.std()

print("median log price=",medianLogP)
print("mean log price=",meanLogP)
print("std log price=",stdLogP)
```

```python
# Transformasi log-10 GrLivArea
luas_tanah=np.log10(house['GrLivArea'])

### Tulis kode Anda di sini
# Median
medianLogLuasT=luas_tanah.median()
# Mean
meanLogLuasT=luas_tanah.mean()
# Std Dev
stdLogLuasT=luas_tanah.std()

print("median log 1 GrLivArea=",medianLogLuasT)
print("mean log 1 GrLivArea=",meanLogLuasT)
print("std log 1 GrLivArea=",stdLogLuasT)
```

```python
# Menghilangkan nilai nol pada GarageArea
luas_tanah_tidak0=house[house['GrLivArea']>0]
luas_tanah_tidak0_logPrice=np.log10(luas_tanah_tidak0['GrLivArea'])

### Tulis kode Anda di sini
# Median
medianLogLuasT=luas_tanah_tidak0_logPrice.median()
# Mean
meanLogLuasT=luas_tanah_tidak0_logPrice.mean()
# Std Dev
stdLogLuasT=luas_tanah_tidak0_logPrice.std()

print("median log 1 GrLivArea=",medianLogLuasT)
print("mean log 1 GrLivArea=",meanLogLuasT)
print("std log 1 GrLivArea=",stdLogLuasT)
```

#### **Penjelasan dan analisis tentang pemrosesan data**

* Median vs Mean: Median (3.1655) sedikit lebih besar dari mean (3.1563). Dalam data yang ditransformasikan secara logaritmik, ini bisa menandakan sedikit skewness (kemiringan) positif pada distribusi asli GrLivArea. Artinya, mungkin ada beberapa rumah dengan luas area tinggal yang sangat besar yang sedikit menarik rata-rata ke atas.
* Standar Deviasi: Standar deviasi yang relatif kecil setelah transformasi logaritma menunjukkan bahwa transformasi tersebut berhasil dalam mengurangi variasi dan membuat data lebih simetris. Ini bermanfaat untuk analisis statistik lebih lanjut.
Kesimpulan:

Secara keseluruhan, analisis ini menunjukkan bahwa luas area tinggal (GrLivArea) memiliki distribusi yang mungkin sedikit miring positif. Transformasi logaritma telah diterapkan untuk mengurangi skewness dan membuat data lebih sesuai untuk analisis statistik. Median dan mean dari data yang ditransformasikan memberikan gambaran tentang luas area tinggal yang umum, sedangkan standar deviasi menunjukkan sebaran data di sekitar rata-rata.

#### **Hipotesis tentang hubungan antar-variabel**

* GarageArea > 0 bertujuan untuk memastikan bahwa hanya rumah-rumah dengan garasi yang memiliki area non-nol yang dianalisis. filter ini memastikan bahwa analisis hanya berfokus pada properti yang relevan, sehingga hasilnya lebih representatif dan valid.

* GrLivArea dan SalePrice:

  * Hubungan: Terdapat korelasi positif yang kuat antara luas ruang tinggal (GrLivArea) dan harga jual rumah (SalePrice).
  * Alasan: Luas ruang tinggal merupakan faktor penting yang dipertimbangkan pembeli properti. Rumah dengan luas ruang tinggal yang lebih besar umumnya menawarkan lebih banyak ruang fungsional, kenyamanan, dan fleksibilitas bagi penghuninya. Oleh karena itu, pembeli bersedia membayar harga yang lebih tinggi untuk rumah dengan luas ruang tinggal yang lebih besar.
  * Implikasi: Peningkatan luas ruang tinggal cenderung diiringi dengan peningkatan harga jual rumah.

* GarageArea dan SalePrice:

  * Hubungan: Terdapat korelasi positif antara luas garasi (GarageArea) dan harga jual rumah (SalePrice), namun hubungan ini tidak sekuat hubungan antara luas ruang tinggal dan harga jual.
  * Alasan: Garasi merupakan fasilitas tambahan yang bernilai bagi banyak pembeli, terutama di daerah dengan iklim yang bervariasi atau di mana kepemilikan kendaraan pribadi merupakan hal yang umum. Garasi memberikan perlindungan untuk kendaraan, ruang penyimpanan tambahan, dan potensi untuk penggunaan lain seperti bengkel atau ruang hobi.
  * Implikasi: Meskipun luas garasi dapat memengaruhi harga jual, pengaruhnya tidak sebesar luas ruang tinggal. Faktor-faktor lain, seperti lokasi, kondisi rumah, dan fitur-fitur lainnya, mungkin memiliki bobot yang lebih besar dalam menentukan harga jual.

* GrLivArea dan GarageArea:

  * Hubungan: Kemungkinan terdapat interaksi antara luas ruang tinggal dan luas garasi dalam memengaruhi harga jual rumah.
  * Alasan: Rumah dengan luas ruang tinggal yang besar dan garasi yang luas mungkin dianggap lebih menarik dan bernilai bagi pembeli, sehingga berpotensi meningkatkan harga jual secara signifikan.
  * Implikasi: Kombinasi luas ruang tinggal dan luas garasi yang ideal dapat menghasilkan nilai jual yang optimal bagi properti.
